## DATA VISUALISATION: 

```python
# Heatmap for correlation analysis
columns_corr = data[['Duration', 'Rating', 'Votes']]
correlation_matrix = columns_corr.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

#### There is very nominal correlation factor amongst any of the numeric variables like votes, rating etc. 
#### This tells us that none of the factors in a movie decides the value of other, in a significant way.

```python
# Histogram to visualise the range of ratings throughout the era
plt.hist(data['Rating'], bins=10, color='pink',edgecolor='black')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Movie Ratings')
plt.show()
```

#### Out of 5659 movies, 1400 of them has got a rating of around 7, while very few movies could actually touch the upper limit of 10.
#### Maximum number of movies have a got a fair rating of more than 5.

```python
# Visualisation of ratings and duration with scatter plot
sns.scatterplot(x='Duration', y='Rating', data=data, color='green')
plt.xlabel('Duration(in minutes)')
plt.ylabel('Rating')
plt.title('Duration vs Rating')
plt.show()
```

#### The scatter plot shows the cluster of movies within the duration of 100-200 minutes, showing that almost all movies lies in this range.
#### There is not much definitive positive or negative correlation, but we do get some prominent outliers. 
#### A movie with maximum duration and a movie with minimum duration, both reflecting a similar rating of around 8. This indicates that duration doesn't channelise how movies get rated.

```python
# Aggregate of all the ratings received in one year, and plotting a time-series plot with it
average_rating_by_year = data.groupby('Year')['Rating'].mean()
plt.plot(average_rating_by_year.index, average_rating_by_year.values)
plt.xlabel("Year")
plt.ylabel("Average Rating")
plt.title("Average Rating of Movies by Year")
plt.show()
```

#### This line graph is much like a time-series vision, which shows the fluctuation of average ratings of movies with time. 
#### There is no monotonic behaviour observed here, the ratings abruptly showed changes in several years.

#### The decade of 2000 saw movies with lowest rating, while as is said, Old is Gold, we can see the highest rated movies in the era of 1940s and 1950s. 

```python
# Box plot for visualising a continuous data, viz duration.
plt.figure(figsize=(8, 6))
sns.boxplot(x=pd.cut(data['Duration'], bins=5), y='Rating', data=data)
plt.title('Comparison of Rating with Duration Ranges')
plt.xlabel('Duration Ranges')
plt.ylabel('Rating')
plt.xticks(rotation=45)
plt.show()
```

#### The highest and the lowest rated movie, both lies in the orange plot of 81 mins to 141 mins range, showing the widest range of ratings in that time span.
#### Movies of 200-260 mins(3- 4 hrs) are maximum in count, followed by 1-2 hr. Accordingly there ratings range not less than 5, but not more than 8 - showing an average level.

### 4. *Present to you, the most popular movie & actor*:

```python
# Summing up all the available actors of the dataset, in one list.
actors = ['Actor 1', 'Actor 2', 'Actor 3']
actor_popularity = {}
# Calculating actor's popularity with the total number of votes they got in their acting history.
for col in actors:
    actor_votes = data.groupby(col)['Votes'].sum().to_dict()
    for actor, votes in actor_votes.items():
        if actor in actor_popularity:
            actor_popularity[actor] += votes
        else:
            actor_popularity[actor] = votes

# Adding popularity values obtained above row-wise in data, for each actor present in that row.           
data['TotalActor_popularity'] = data[['Actor 1', 'Actor 2', 'Actor 3']].apply(lambda row: sum(actor_popularity[actor] for actor in row), axis=1)
```

```python
data
```

```python
# Most Popular actor
popular_actor = max(zip(actor_popularity.values(), actor_popularity.keys()))[1]  
print(popular_actor)
```

```python
# Most Popular Movie
data[data['TotalActor_popularity']==data['TotalActor_popularity'].max()]
```

#### We see the popularity of a movie, by the popularity of the actors in it.
#### An actor is considered popular by the total number of votes he got in his journey.
#### This popularity of movie is being evaluated by the total votes of each actor in the journey of their acting, and is being added for each movie.
#### To conclude, Bollywood has: 
#### The action-adventure genre film 'Thugs of Hindustan' as the most popular movie, and
####  Aamir Khan as the most popular actor.
